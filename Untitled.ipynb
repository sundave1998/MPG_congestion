```python
NPG_plot_accuracies = np.load("NPG_Proposed_runs_01.npy")
PG_plot_accuracies = np.load("PG_Ding_runs.npy")
PGA_plot_accuracies = np.load("PGA_Leonardos_runs.npy")





```

```python

clrs = sns.color_palette("husl", 3)

fig1 = plt.figure(figsize=(6,4))

pmean = list(map(statistics.mean, zip(*NPG_plot_accuracies)))
pstdv = list(map(statistics.stdev, zip(*NPG_plot_accuracies)))
ax = sns.lineplot(piters, pmean, color = clrs[0],label= 'NPG Mean L1-accuracy')
ax.fill_between(piters, np.subtract(pmean,pstdv), np.add(pmean,pstdv), alpha=0.3, facecolor=clrs[0],label="NPG 1-standard deviation")


pmean = list(map(statistics.mean, zip(*PG_plot_accuracies)))
pstdv = list(map(statistics.stdev, zip(*PG_plot_accuracies)))
ax = sns.lineplot(piters[:len(PG_plot_accuracies[i])], pmean, color = clrs[1],label= 'PG Mean L1-accuracy')
ax.fill_between(piters[:len(PG_plot_accuracies[i])], np.subtract(pmean,pstdv), np.add(pmean,pstdv), alpha=0.3, facecolor=clrs[1],label="PG 1-standard deviation")


pmean = list(map(statistics.mean, zip(*PGA_plot_accuracies)))
pstdv = list(map(statistics.stdev, zip(*PGA_plot_accuracies)))
ax = sns.lineplot(piters[:len(PGA_plot_accuracies[i])], pmean, color = clrs[2],label= 'PGA Mean L1-accuracy')
ax.fill_between(piters[:len(PGA_plot_accuracies[i])], np.subtract(pmean,pstdv), np.add(pmean,pstdv), alpha=0.3, facecolor=clrs[2],label="PGA 1-standard deviation")



ax.legend()
plt.grid(linewidth=0.6)
plt.gca().set(xlabel='Iterations',ylabel='L1-accuracy', title='Policy Gradient: agents = {}, runs = {}'.format(8, len(NPG_plot_accuracies)))
plt.show()
fig1.savefig('Compare_avg_runs_n{}.png'.format(8),bbox_inches='tight')
```
